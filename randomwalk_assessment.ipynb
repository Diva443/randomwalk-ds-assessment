```python
#Q:7 Bucket the avg rating books and plot with bar graph

bins=[0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0] #given 11 buckets
label= [f'{x:.1f}-{x+0.5:.1f}' for x in bins[:-1]]  # for rounding 0.5 decimal with ratings
```

```python
#creating column rating_bucket for our buckets
data['rating_bucket']= pd.cut(data['average_rating'],bins=bins, labels=label,right=False)
bucket_count= data['rating_bucket'].value_counts().sort_index()
pd.DataFrame(bucket_count)
```

```python
#ploting
plt.bar(bucket_count.index,bucket_count.values)
plt.xlabel("Rating_bucket")
plt.ylabel("Number of books")
plt.title('Number of books in each RatingBuckets')
plt.show()
```
